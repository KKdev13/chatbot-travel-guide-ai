This mini project -serves as a proof of concept- harnessing the potential of OpenAI's API, to create an AI-powered travel guide for the culturally rich city of Paris.

The chatbot acts as a virtual Parisian expert, delivering valuable insights into the city's iconic landmarks and hidden treasures. The AI will respond intelligently to a set of common questions, providing a more engaging and immersive travel planning experience for the clientele of Peterman Reality Tours.

The ultimate aspiration is a user-friendly, AI-driven travel guide that significantly enhances the exploration of Paris. Users will be able to pre-define their questions and receive well-informed answers from the AI, providing a seamless and intuitive travel planning process.

```python

import os
from openai import OpenAI

# Define the model to use
model = "gpt-4o-mini"

# Define the client
client = OpenAI()

# Create the conversation with system message and user questions
conversation = [
    {"role": "system", "content": "You are a knowledgeable Parisian travel guide expert. You provide accurate, concise information about Paris landmarks, attractions, and cultural insights. You are friendly and helpful, specializing in helping tourists explore the city."},
    {"role": "user", "content": "How far away is the Louvre from the Eiffel Tower (in miles) if you are driving?"},
    {"role": "user", "content": "Where is the Arc de Triomphe?"},
    {"role": "user", "content": "What are the must-see artworks at the Louvre Museum?"}
]

# Generate responses for each question
responses = []
for message in conversation[1:]: #skip the system message
    #Create a temp converstion for this query
    temp_conversation = [conversation[0], message]

    #Get response from OpenAI
    response = client.chat.completions.create(
        model = model,
        messages = temp_conversation,
        temperature = 0.0,
        max_tokens = 100
    )

    #Extract the response content
    assistant_response = response.choices[0].message.content

    #Append to responses list
    responses.append(assistant_response)

# Build the full conversation with all responses
full_conversation = [conversation[0]]  # Start with system message

# Interleave questions and responses
for i, user_msg in enumerate(conversation[1:]):
    full_conversation.append(user_msg)  # Add user question
    full_conversation.append({"role": "assistant", "content": responses[i]})  # Add assistant response

# The final conversation list
conversation = full_conversation

# Print the conversation to see the results
for message in conversation:
    print(f"{message['role']}: {message['content']}\n")
```
